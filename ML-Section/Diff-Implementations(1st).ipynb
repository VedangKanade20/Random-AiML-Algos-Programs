```python
# üìå Required Libraries
import pandas as pd
import numpy as np
from sklearn.datasets import load_iris
from sklearn.feature_selection import SelectKBest, f_classif
from sklearn.preprocessing import MinMaxScaler, PowerTransformer
from sklearn.decomposition import PCA

# üìå Load Dataset
data = load_iris()
X = pd.DataFrame(data.data, columns=data.feature_names)
y = data.target
print("Original Shape:", X.shape)

# -------------------------------------------------------
# 1Ô∏è‚É£ Feature Extraction ‚Äì Create new feature
# -------------------------------------------------------
X['sepal_ratio'] = X['sepal length (cm)'] / X['sepal width (cm)']
print("After Feature Extraction:", X.shape)

# -------------------------------------------------------
# 2Ô∏è‚É£ Feature Selection ‚Äì Keep only 4 best features
# -------------------------------------------------------
selector = SelectKBest(score_func=f_classif, k=4)
X_selected = selector.fit_transform(X, y)
print("After Feature Selection:", X_selected.shape)

# -------------------------------------------------------
# 3Ô∏è‚É£ Normalization ‚Äì Scale values between 0 and 1
# -------------------------------------------------------
scaler = MinMaxScaler()
X_normalized = scaler.fit_transform(X_selected)

# -------------------------------------------------------
# 4Ô∏è‚É£ Transformation ‚Äì Make data more Gaussian-like
# -------------------------------------------------------
transformer = PowerTransformer()
X_transformed = transformer.fit_transform(X_normalized)

# -------------------------------------------------------
# 5Ô∏è‚É£ PCA ‚Äì Reduce dimensions to 2 principal components
# -------------------------------------------------------
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_transformed)
print("After PCA:", X_pca.shape)
```
